# Risk factors for Colombian High School students at the ICFES Exam

## Context
The Colombian Institute for the Promotion of Higher Education (ICFES) has been designing and managing the state examinations for admission to higher education since 1968 for students who have completed secondary education and are planning to enter a university or college.

This exam is a standardized test similar to the American SAT. The test is administered before graduation in Colombian high schools. The test is thoroughly developed and published by the ICFES.

Several social and economic factors may affect students' performance at the ICFES exam. This project aims to build a model to identify High School students located in the Atlantico region with higher probabilities of having poor performance, so local schools can take action on those students and give them special attention and resources.

### Load and explore the data

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')

# Load the data files
data_2019 = pd.read_csv("../data/Saber_11__2019-2.csv")
data_2020 = pd.read_csv("../data/Saber_11__2020-2.csv")

# Check consistency between dataframeÂ´s columns
print(data_2019.shape)
print(data_2020.shape)
```

Since data_2019 has an additional column, it's necessary to drop that extra column

```python
# Get a list for each of the columns
columns_2019 = data_2019.columns
columns_2020 = data_2020.columns

# Get the common columns among both lists
common_columns = [column for column in columns_2019 if column in columns_2020]

# Keep only common columns
data_2019 = data_2019[common_columns]

# Join both dataframes
data = pd.concat([data_2019, data_2020])

# Explore the columns:
data.info()
```

### Filter by students located in  Atlantico region
Since this project aims to detect risk factors to students located in the Atlantico region, the dataset will be reduced to these population.

```python
data = data[data["COLE_DEPTO_UBICACION"] == "ATLANTICO"]
data.shape
```

### Keep relevant columns
Not all the 81 columns can bring useful information. Only the most relevant will remain.

```python
# Create a new column with the age of students
birth_year = [datetime.strptime(i, "%m/%d/%Y %I:%M:%S %p" ).year for i in data["ESTU_FECHANACIMIENTO"]]
icfes_year = [datetime.strptime(str(i), "%Y%d").year for i in data["PERIODO"]]
age = [icfes-birth for icfes,birth in zip(icfes_year, birth_year)]
data["EDAD"] = age


# Filter by relevant columns
relevant_columns = ["ESTU_GENERO", "EDAD", "COLE_NATURALEZA", "COLE_CARACTER",\
"COLE_AREA_UBICACION", "COLE_MCPIO_UBICACION", "COLE_NOMBRE_ESTABLECIMIENTO", "FAMI_ESTRATOVIVIENDA",\
"FAMI_TIENEINTERNET", "FAMI_TIENECOMPUTADOR", "ESTU_DEDICACIONLECTURADIARIA",\
"ESTU_HORASSEMANATRABAJA","PUNT_LECTURA_CRITICA",\
"DESEMP_LECTURA_CRITICA", "PUNT_MATEMATICAS", "DESEMP_MATEMATICAS", \
"PUNT_C_NATURALES", "DESEMP_C_NATURALES", "PUNT_SOCIALES_CIUDADANAS",\
"DESEMP_SOCIALES_CIUDADANAS","PUNT_INGLES", "PUNT_GLOBAL"]

data_lite = data[relevant_columns].copy()
```

### Handling outliers on the "Age" column

```python
# Explore ande remove outliers on the "AGE" column
print("Max age:", data_lite["EDAD"].max())
print("Min age:", data_lite["EDAD"].min())
print("")

credible_ages = [i for i in data_lite["EDAD"] if 13<i<80]
avg_ages = round(np.mean(credible_ages))
print("AVG(age):", avg_ages)
print("")

# Replace ages avobe 80 years old
data_lite["EDAD"].values[data_lite["EDAD"] > 80] = avg_ages
print("New max age:", data_lite["EDAD"].max())

# Replace ages below 13 years old
data_lite["EDAD"].values[data_lite["EDAD"] < 13] = avg_ages
print("New min age:", data_lite["EDAD"].min())
```

### Handling missing values

Missing values will be handled as follows:
- Records with missing values on the COLE_CARACTER column will be dropped since there isn't a convenient way to fill them.
- Missing values on the FAMI_ESTRATOVIVIENDA column will be filled with the School mode.
- Missing values on the FAMI_TIENEINTERNET column will be filled with the School mode.
- Missing values on the FAMI_TIENECOMPUTADOR column will be filled with the School mode.
- Missing values on the ESTU_DEDICACIONLECTURADIARIA column will be filled with the School mode.
- Missing values on the ESTU_HORASSEMANATRABAJA column will be filled with the School mode.

```python
# Records with missing values on the COLE_CARACTER column will be dropped since there isn't a convenient way to fill those
data_lite = data_lite.dropna(subset = ["COLE_CARACTER"])

# Missing values on the FAMI_ESTRATOVIVIENDA column will be filled with the School mode
f = lambda x: x.median() if np.issubdtype(x.dtype, np.number) else x.mode().iloc[0]

data_lite["FAMI_ESTRATOVIVIENDA"].fillna(data_lite.groupby("COLE_NATURALEZA")\
                                         ["FAMI_ESTRATOVIVIENDA"].transform(f),inplace=True)

# Missing values on the FAMI_TIENEINTERNET column will be filled with the School mode
data_lite["FAMI_TIENEINTERNET"].fillna(data_lite.groupby("COLE_NATURALEZA")\
                                       ["FAMI_TIENEINTERNET"].transform(f),inplace=True)

# Missing values on the FAMI_TIENECOMPUTADOR column will be filled with the School mode
data_lite["FAMI_TIENECOMPUTADOR"].fillna(data_lite.groupby("COLE_NATURALEZA")\
                                                    ["FAMI_TIENECOMPUTADOR"].transform(f),inplace=True)

# Missing values on the ESTU_DEDICACIONLECTURADIARIA column will be filled with the School mode
data_lite["ESTU_DEDICACIONLECTURADIARIA"].fillna(data_lite.groupby("COLE_NATURALEZA")\
                                                 ["ESTU_DEDICACIONLECTURADIARIA"].transform(f),inplace=True)

# Missing values on the ESTU_HORASSEMANATRABAJA column will be filled with the School mode
data_lite["ESTU_HORASSEMANATRABAJA"].fillna(data_lite.groupby("COLE_NATURALEZA")\
                                            ["ESTU_HORASSEMANATRABAJA"].transform(f),inplace=True)
```

### When is a student considered as "at risk"?

A score above 250 is considered positive (**infobae.com**). Therefore, students with a score lower than 150 will be considered as "students at risk".   
A new column will be created to indicate if the student is at risk:

```python
risk_criteria = 150
risk_flag = np.array([1 if i < risk_criteria else 0 for i in data_lite["PUNT_GLOBAL"]])

df = data_lite.copy()
df["RISK_FLAG"] = risk_flag

cols_to_drop = ["PUNT_LECTURA_CRITICA", "DESEMP_LECTURA_CRITICA", "PUNT_MATEMATICAS",
               "DESEMP_MATEMATICAS", "PUNT_C_NATURALES", "DESEMP_C_NATURALES", "PUNT_GLOBAL",
               "PUNT_SOCIALES_CIUDADANAS", "DESEMP_SOCIALES_CIUDADANAS", "PUNT_INGLES",
               "COLE_NOMBRE_ESTABLECIMIENTO", "COLE_MCPIO_UBICACION"]
                
df_model = df.drop(cols_to_drop, axis=1)

# Creating a dummy dataframe
df_dummy = pd.get_dummies(df_model, drop_first = True)
```

### Dividing the data into train and test sets

The data will be balanced to have the same number of students at risk and students not at risk. 75% of the data will be used to train the models, and the remaining 25% to test them.

```python
# Importing train test split function
from sklearn.model_selection import train_test_split

df_risk = df_dummy[df_dummy["RISK_FLAG"] == 1]
df_not_risk = df_dummy[df_dummy["RISK_FLAG"] == 0]

df_not_risk = df_not_risk.sample(n=int(len(df_risk)*1.2), random_state = 1111)


df_balanced = pd.concat([df_not_risk, df_risk])
df_balanced = df_balanced.sample(n=len(df_risk)+len(df_not_risk), random_state = 1111)


y = df_balanced["RISK_FLAG"].values
X = df_balanced.drop("RISK_FLAG", axis = 1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1111, stratify = y)
```

### Initializing the models
The models to consider on this project are:
- Logistic Regression
- Decision Tree
- KNN
- Random Forest
- Ada Boost
- Voting Classifier

```python
# Importing the models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import VotingClassifier

# Create the models
lr = LogisticRegression(random_state = 1111)
knn = KNN(n_neighbors = 6)
dt = DecisionTreeClassifier(criterion = 'gini', random_state = 1111, max_depth = round(len(df_dummy)**(1/2)))
rfc = RandomForestClassifier(n_estimators = 500 , min_samples_leaf = 0.12, random_state = 1111)
ada = AdaBoostClassifier(n_estimators = 50, random_state = 1111)
```

### Evaluating the models
The metrics to evaluate the models are accuracy and ROC score.

```python
# Importing the metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

classifiers = [("Logistic Regression", lr), ("K Nearest Neighbors", knn),("Classification Tree", dt),\
              ("Random Forest", rfc), ("Ada Boost", ada)]

vc = VotingClassifier(estimators = classifiers, voting='soft')

for clf_name, clf in classifiers:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    y_pred_proba = clf.predict_proba(X_test)[:,1]
    roc_score = roc_auc_score(y_test, y_pred_proba)
    
    print('{:s} accuracy: {:.3f}'.format(clf_name, accuracy_score(y_test, y_pred)))
    print('{:s} ROC score: {:.3f}'.format(clf_name, roc_score))
    print('')
    
vc.fit(X_train, y_train)
y_pred_vc = vc.predict(X_test)
y_pred_proba_vc = vc.predict_proba(X_test)[:,1]
roc_score_vc = roc_auc_score(y_test, y_pred_proba_vc)
print('Voting Classifier accuracy: {:.3f}'.format(accuracy_score(y_test, y_pred_vc)))
print('Voting Classifier ROC score: {:.3f}'.format(roc_score_vc))
```

### ROC Curves

```python
# Importing libraries to plot the ROC curve
from sklearn.metrics import plot_roc_curve

sns.set()
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (18,12))

classifiers_roc = [("Logistic Regression", lr), ("K Nearest Neighbors", knn),("Classification Tree", dt),\
              ("Random Forest", rfc), ("Ada Boost", ada), ("Voting classifier", vc)]

for (clf_name, clf), idx in zip(classifiers_roc, axes.flatten()):
    clf.fit(X_train, y_train)
    plot_roc_curve(clf, X_test, y_test, ax = idx)
```

### Confusion Matrices

```python
from sklearn.metrics import confusion_matrix

sns.set()
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (18,12))

classifiers_matrix = [("Logistic Regression", lr), ("K Nearest Neighbors", knn),("Classification Tree", dt),\
              ("Random Forest", rfc), ("Ada Boost", ada), ("Voting classifier", vc)]


for (clf_name, clf), idx in zip(classifiers_matrix, axes.flatten()):

    conf_matrix = confusion_matrix(y_true = y_test, y_pred = clf.predict(X_test))
    idx.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
    
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            idx.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
        
    idx.set_xlabel('Predictions', fontsize=12)
    idx.set_ylabel('Actuals', fontsize=12)
    idx.set_title(clf_name, fontsize=12)
```

### Conclusions about the models
